#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

import matplotlib.pyplot as plt

# List of features to plot against Steps
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
titles = ['Minutes Lightly Active vs Steps', 'Minutes Very Active vs Steps', 'Minutes Sedentary vs Steps']

# Create scatter plots
plt.figure(figsize=(15, 5))
for i, feature in enumerate(features):
    plt.subplot(1, 3, i+1)
    plt.scatter(fitbit[feature], fitbit['Steps'], alpha=0.5)
    plt.title(titles[i])
    plt.xlabel(feature)
    plt.ylabel('Steps')

plt.tight_layout()
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

plt.style.use('classic')
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

bin_sizes = [10, 20, 50]
for ax, bins in zip(axes, bin_sizes):
    ax.hist(titanic['Age'].dropna(), bins=bins, edgecolor='black')
    ax.set_title(f'Age Distribution with {bins} bins')
    ax.set_xlabel('Age')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
# your code here

titanic_num = titanic.select_dtypes(include=['float64', 'int64'])
corr_matrix = titanic_num.corr()

# Plotting with Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
