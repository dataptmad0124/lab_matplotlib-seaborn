#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100)
y1 = x**2
y2 = np.exp(x)

# Plot
fig, axs = plt.subplots(2, figsize=(8, 10))

# Subplot 1: Normal scale
axs[0].plot(x, y1, color='blue', label='$y = x^2$')
axs[0].plot(x, y2, color='red', label='$y = e^x$')
axs[0].set_title('Subplot with Normal Scale')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# Subplot 2: Logarithmic scale
axs[1].plot(x, y1, color='blue', label='$y = x^2$')
axs[1].plot(x, y2, color='red', label='$y = e^x$')
axs[1].set_title('Subplot with Logarithmic Scale')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_yscale('log')

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# your code here

# Calcular el promedio de pasos por mes
avg_steps_by_month = fitbit.groupby('Months_encoded')['Steps'].median()

# Calcular el promedio de pasos por d√≠a de la semana
avg_steps_by_weekday = fitbit.groupby('Days_encoded')['Steps'].median()

# Configurar el tama√±o de la figura
plt.figure(figsize=(12, 4))

# Gr√°fico 1: Promedio de pasos por mes
plt.subplot(1, 2, 1)
plt.bar(avg_steps_by_month.index, avg_steps_by_month.values)
plt.title('Median Steps by Month')
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Gr√°fico 2: Promedio de pasos por d√≠a de la semana
plt.subplot(1, 2, 2)
plt.bar(avg_steps_by_weekday.index, avg_steps_by_weekday.values)
plt.title('Median Steps by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Median Steps')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

# Guardar la figura en la carpeta 'figures'
plt.savefig('../Figures/steps.png')

plt.show()

```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

# Crear una lista de colores para cada gr√°fico
colors = ['blue', 'green', 'red']

# Configurar el tama√±o de la figura
plt.figure(figsize=(15, 5))

# Iterar sobre las caracter√≠sticas y trazar los gr√°ficos de dispersi√≥n
for i, feature in enumerate(features):
    plt.subplot(1, 3, i+1)
    plt.scatter(fitbit[feature], fitbit['Steps'], color=colors[i], alpha=0.5)
    plt.title(f'{feature} vs Steps')
    plt.xlabel(feature)
    plt.ylabel('Steps')

# Ajustar el dise√±o de los subgr√°ficos
plt.tight_layout()

# Mostrar los gr√°ficos
plt.show()
```
