```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, axs = plt.subplots(nrows=1,ncols=2, figsize=(12,4))
fitbit_m = fitbit.loc[:, ["Steps","Months_encoded"]]
fitbit_m = fitbit_m.groupby(by="Months_encoded").median()
plt.sca(axs[0])
axs[0].plot(fitbit_m.index.values, fitbit_m["Steps"])
axs[0].set_title("Month encoded")
plt.xticks(rotation="vertical")

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fitbit_w = fitbit.loc[:, ["Steps",'Work_or_Weekend']]
fitbit_w = fitbit_w.groupby(by='Work_or_Weekend').median()
plt.sca(axs[1])
axs[1].plot(fitbit_m.index.values, fitbit_m["Steps"])
axs[1].set_title("Work or Weekend")
plt.xticks(rotation="vertical")

plt.savefig("./fitbit_median.png", format="png" )
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
#your code here
data = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
plt.figure(figsize=(15, 5))
for i, data in enumerate(data,1):
    plt.subplot(1, 3, i)
    grouped_data = fitbit.groupby(data)['Steps'].median().reset_index()
    plt.plot(grouped_data[data], grouped_data['Steps'])
    plt.title(f'{data}')
    

plt.show()

```
