#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# Datos para las nuevas funciones
x = np.arange(1, 100)
y1 = x ** 2
y2 = np.exp(x)

# Crear una figura y un arreglo de ejes con 1 fila y 2 columnas, ajustando el tamaño de la figura
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Primer gráfico: escala normal
axes[0].plot(x, y1, label='y = x^2', color='blue', linewidth=2)
axes[0].plot(x, y2, label='y = e^x', color='green', linewidth=2)
axes[0].set_title('Normal Scale', fontsize=16, fontweight='bold')
axes[0].set_xlabel('x', fontsize=14)
axes[0].set_ylabel('y', fontsize=14)


# Segundo gráfico: escala logarítmica
axes[1].plot(x, y1, label='y = x^2', color='blue', linewidth=2)
axes[1].plot(x, y2, label='y = e^x', color='green', linewidth=2)
axes[1].set_title('Logarithmic Scale', fontsize=16, fontweight='bold')
axes[1].set_xlabel('x', fontsize=14)
axes[1].set_ylabel('y', fontsize=14)
axes[1].set_yscale('log')


# Ajustar el diseño para evitar superposición
plt.tight_layout()
plt.show()

```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4

* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED


# Convertir la columna 'Fecha' a datetime
fitbit['Date'] = pd.to_datetime(fitbit['Date'])

# Calcular los pasos medios por mes
steps_by_month = fitbit.groupby('Months_encoded')['Steps'].mean()

# Comparar la actividad entre los días laborables y los fines de semana
steps_by_day_type = fitbit.groupby('Work_or_Weekend')['Steps'].mean()

# Crear visualizaciones
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Gráfico de pasos medios por mes
axes[0].bar(steps_by_month.index, steps_by_month.values, color='blue')
axes[0].set_title('Average Steps per Month', fontsize=16)
axes[0].set_xlabel('Month', fontsize=12)
axes[0].set_ylabel('Average Steps', fontsize=12)
axes[0].set_xticks(steps_by_month.index)
axes[0].set_xticklabels(steps_by_month.index)

# MEDIAN STEPS BY WORK_OR_WEEKEND

axes[1].bar(steps_by_day_type.index, steps_by_day_type.values, color='green')
axes[1].set_title('Average Steps by Type of Day', fontsize=16)
axes[1].set_xlabel('Type of Day', fontsize=12)
axes[1].set_ylabel('Average Steps', fontsize=12)
axes[1].set_xticks(steps_by_day_type.index)
axes[1].set_xticklabels(['Laborable day', 'Weekend'])

# Ajustar el diseño para evitar superposición
plt.tight_layout()

# Guardar las figuras en la carpeta 'figures'
fig.savefig(r"C:\Users\hp\Documents\M2\lab_matplotlib_seaborn-main\lab_matplotlib_seaborn-main\data")

# Mostrar los gráficos
plt.show()

```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# Columnas que queremos graficar
columns = [
    ("Minutes Lightly Active", "Steps"),
    ("Minutes Very Active", "Steps"),
    ("Minutes Sedentary", "Steps")
]

# Crear una figura y un arreglo de ejes con 1 fila y 3 columnas
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Títulos para los gráficos
titles = [
    "Minutes Lightly Active vs Steps",
    "Minutes Very Active vs Steps",
    "Minutes Sedentary vs Steps"
]

# Crear los scatter plots
for i, (x_col, y_col) in enumerate(columns):
    axes[i].scatter(fitbit[x_col], fitbit[y_col], alpha=0.5)
    axes[i].set_title(titles[i], fontsize=14, fontweight='bold')
    axes[i].set_xlabel(x_col, fontsize=12)
    axes[i].set_ylabel(y_col, fontsize=12)
    axes[i].grid(True)

# Ajustar el diseño para evitar superposición
plt.tight_layout()
# Mostrar los gráficos
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
# Seleccionar solo las columnas numéricas para la matriz de correlación
numeric_columns = titanic.select_dtypes(include=['int64', 'float64']).columns
titanic_numeric = titanic[numeric_columns]

# Calcular la matriz de correlación
corr_matrix = titanic_numeric.corr()

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Crear el mapa de calor utilizando Seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .7})

# Añadir título
plt.title('Correlation Matrix of Numeric Variables', fontsize=16)

# Mostrar el gráfico
plt.show()
```
