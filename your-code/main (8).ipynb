#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# Definimos y al cuadrado y representamos.

y2 = [math.exp(i) for i in x]

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11, 5))
ax[0].plot(x, y)
ax[1].plot(x, y2)

ax[0].set_xscale('linear')
ax[1].set_yscale('log')

ax[0].set_title("Eje x, y")
ax[1].set_title("Eje x, y2")

plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# Hemos aadido una leyenda a la grfica

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11, 5))
ax[0].plot(x, y)
ax[1].plot(x, y2)

ax[0].set_xscale('linear')
ax[1].set_yscale('log')

ax[0].set_title("Eje x, y")
ax[1].set_title("Eje x, y2")

ax[0].legend(["linear"])
ax[1].legend([]"log"])

plt.show()
```

From the Fitbit data, we want to visually understand:
How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# Damos tamao al layout y decidimos que haya una fila y dos columnas
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# MEDIANA de STEPS por MONTH_ENCODED

fitbit_m = fitbit.loc[:, ['Steps','Months_encoded']]
fitbit_m = fitbit_m.groupby(by='Months_encoded').median()

plt.sca(ax[0])
ax[0].plot(fitbit_m.index.values, fitbit_m['Steps'])


# MEDIANA de STEPS por WORK_OR_WEEKEND

fitbit_w = fitbit.loc[:, ['Steps','Work_or_Weekend']]
fitbit_w = fitbit_w.groupby(by='Work_or_Weekend').median()

plt.sca(ax[1])
ax[1].plot(fitbit_w.index.values, fitbit_w['Steps'])

plt.xticks(rotation = 'vertical')

plt.savefig('../saved_figures/figura.png')

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# Esta mal, tengo que hacer un bucle y meter estas tres grficas en una.

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))

fitbit_mla = fitbit.loc[:, ['Steps', 'Minutes Lightly Active']]
fitbit_mva = fitbit.loc[:, ['Steps','Minutes Very Active']]
fitbit_ms = fitbit.loc[:, ['Steps', 'Minutes Sedentary']]


plt.sca(ax[0])
plt.scatter(fitbit_mla['Minutes Lightly Active'], fitbit_mla['Steps'], color='red')
plt.title('Minutes Lightly Active')
plt.xlabel('Minutes')
plt.ylabel('Steps')


plt.sca(ax[1])
plt.scatter(fitbit_mva['Minutes Very Active'], fitbit_mva['Steps'], color='blue')
plt.title('Minutes Very Active')
plt.xlabel('Minutes')
plt.ylabel('Steps')


plt.sca(ax[2])
plt.scatter(fitbit_ms['Minutes Sedentary'], fitbit_ms['Steps'], color='green')
plt.title('Minutes Sedentary')
plt.xlabel('Minutes')
plt.ylabel('Steps')

plt.show()
```

```python
# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

colors = ['red', 'blue', 'green']

features = [
    ('Minutes Lightly Active', 'red'),
    ('Minutes Very Active', 'blue'),
    ('Minutes Sedentary', 'green')]

# Iterar sobre los datos y graficar cada uno
for idx, (feature, color) in enumerate(features):
    plt.scatter(fitbit[feature], fitbit['Steps'], color=color, label=feature)

plt.title('Scatter Plots of Activity vs Steps')
plt.xlabel('Minutes')
plt.ylabel('Steps')

plt.legend()

plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# Represento tres graficas en el que indico en cada una de ellas el numero de barras que yo quiero.
# El dato principal es la edad.

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

bins_values = [10, 20, 50]

for i, bins in enumerate(bins_values):
    
    sns.histplot(data=titanic, x='Age', bins=bins, ax=axes[i], color='skyblue')
    
    axes[i].set_title(f'Distribution of Age (bins={bins})')
    axes[i].set_xlabel('Age')
    axes[i].set_ylabel('Frequency')

# Ajustar el espacio entre los subplots
plt.tight_layout()

plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
# Seleccionar solo las columnas numricas del conjunto de datos, porque si coges string te da todo fallo.

columnas_numericas = titanic.select_dtypes(include=['float64', 'int64'])

correlacion_matrix = columnas_numericas.corr()

plt.figure(figsize=(14, 10))

# Plotear la matriz de correlacin utilizando Seaborn
sns.heatmap(correlacion_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Correlacin Matrix')

plt.show()
```
