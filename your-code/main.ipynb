#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0, 100)
y1 = x**2
y2 = np.exp(x)


fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6)) #11,6 porque sí, se puede ir probando para ver la dimensión
ax1.plot(x, y1)
ax1.plot(x, y2,)

ax1.set_xscale('linear') #esto cambia los ejes NO los datos
ax1.set_yscale('log')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
#avg=(fitbit['Steps'].mean())
avg_month=fitbit.groupby('Months_encoded')['Steps'].mean()
x = avg_month.index
y1 = avg_month.values



fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4)) 
ax1.plot(x, y1)
ax1.plot(x, y1)
ax1.set_title('Average steps per month')
ax1.set_xlabel('Months')
ax1.set_ylabel('Steps')
plt.xticks(avg_month.index) 

plt.savefig('figures/challenge1.jpg')
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here


median_month=fitbit.groupby('Months_encoded')['Steps'].median()
x = median_month.index
y1 = median_month.values



fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4)) 
ax1.plot(x, y1)
ax1.plot(x, y1)
ax1.set_title('Median steps per month')
ax1.set_xlabel('Months')
ax1.set_ylabel('Steps')


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

work_days=fitbit[fitbit['Days_encoded'] <= 4]
weekend=fitbit[fitbit['Days_encoded'] >= 5]
median_work=work_days.groupby('Days_encoded')['Steps'].median()
median_weekend=weekend.groupby('Days_encoded')['Steps'].median()
x = median_work.index
y1 = median_work.values, weekend



fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4)) 
ax1.plot(median_work.index, median_work.values, marker='o', linestyle='-', color='b')
ax2.plot(median_weekend.index, median_weekend.values, marker='o', linestyle='-', color='r')

ax1.set_title('Median steps workdays')
ax1.set_xlabel('day')
ax1.set_ylabel('Steps')
ax2.set_title('Median steps weekend')
ax1.set_xlabel('day')
ax1.set_ylabel('Steps')

plt.savefig('figures/challenge2.jpg')
```
